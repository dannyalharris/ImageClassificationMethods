#### Flip

```python
for i in range(len(buerste)):
        print([i])
        print(buerste[i])
        j = i + 800
        k = i + 1200
        im = cv2.imread(str(buerste[i]))
        # Make horizontal flip
        im_h = cv2.flip(im, 0)
        # Make vertical flip
        im_v = cv2.flip(im, 1)
        # Save vertical and horizontal image
        cv2.imwrite(savepath_buerste +str(j)+".png", im_h)
        cv2.imwrite(savepath_buerste +str(k)+".png", im_v)
```

```python
for i in range(len(kamm)):
        print([i])
        print(kamm[i])
        j = i + 800
        k = i + 1200
        im = cv2.imread(str(kamm[i]))
        # Make horizontal flip
        im_h = cv2.flip(im, 0)
        # Make vertical flip
        im_v = cv2.flip(im, 1)
        # Save vertical and horizontal image
        cv2.imwrite(savepath_kamm +str(j)+".png", im_h)
        cv2.imwrite(savepath_kamm +str(k)+".png", im_v)
```

#### Gaussian blur

```python
for i in range(len(buerste)):
        print([i])
        print(buerste[i])
        j = i+1600
        im = cv2.imread(str(buerste[i]))
        # Gaussian blur with kernel value equal to 5
        im_g = cv2.GaussianBlur(im,(5,5), 0)
        # Save Gaussian blur image
        cv2.imwrite(savepath_buerste +str(j)+".png", im_g)
```

```python
for i in range(len(kamm)):
        print([i])
        print(kamm[i])
        j = i+1600
        im = cv2.imread(str(kamm[i]))
        # Gaussian blur with kernel value equal to 5
        im_g = cv2.GaussianBlur(im,(5,5), 0)
        # Save Gaussian blur image
        cv2.imwrite(savepath_kamm +str(j)+".png", im_g)
```

#### Histogram equalization

```python
for i in range(len(buerste)):
        print([i])
        print(buerste[i])
        j = i+2400
        im = cv2.imread(str(buerste[i]),0)
        im_he = cv2.equalizeHist(im)
        cv2.imwrite(savepath_buerste +str(j)+".png", im_he)
```

```python
for i in range(len(kamm)):
        print([i])
        print(kamm[i])
        j = i+2400
        im = cv2.imread(str(kamm[i]),0)
        im_he = cv2.equalizeHist(im)
        cv2.imwrite(savepath_kamm +str(j)+".png", im_he)
```

#### Contrast Limited Adaptive Histogram Equalization

```python
for i in range(len(buerste)):
        print([i])
        print(buerste[i])
        j = i+2800
        # Create CLAHE with contrast limit of 2 and image block tile size of 8x8
        clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
        im = cv2.imread(str(buerste[i]),0)
        im_clahe = clahe.apply(im)
        cv2.imwrite(savepath_buerste +str(j)+".png", im_clahe)
```
